# Gas in cosmological simulations

For this task we are going to use a dataset of gas particles of a cosmological simulation. To do so, the cosmological code [**GADGET2**](http://www.mpa-garching.mpg.de/gadget/) was used, implementing the hydrodynamical scheme [**SPH** (Smoothed Particle Hydrodynaics)](http://en.wikipedia.org/wiki/Smoothed-particle_hydrodynamics). This cosmological simulation is consistent with the [WMAP7 cosmology](https://letterstonature.wordpress.com/2010/01/29/wmap-7-cosmological-parameter-set/), and it has a comoving lenght of $20\ h^{-1}Mpc$ per side (a cubic box), with a resolution of $64^3=262144$ particles of gas.

The information of the gas particles can be downloaded from [here](https://raw.githubusercontent.com/sbustamante/ComputationalMethods/master/data/SPH_gas.dat). The order of the columns is:

1. ID of each particle.
2. X coordinate [$pc$ units].
3. Y coordinate [$pc$ units].
4. Z coordinate [$pc$ units].
5. X velocity component [$km/s$ units].
6. Y velocity component [$km/s$ units].
7. Z velocity component [$km/s$ units].
8. Mass of each particle. Same for all [$10^{10} M_{\odot}$ units].
9. Internal Energy.
10. Density.
11. Pressure.
12. Temperature.
13. Redshift.
14. Cosmological time [units in age of the universe].
