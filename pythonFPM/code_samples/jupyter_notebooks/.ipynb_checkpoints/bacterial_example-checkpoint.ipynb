Explore two-parameter model for enzyme activity.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.close('all')

#%% (a) Generate model data for A=1, tau=1, 0<t<2
# Initialize parameters.
time = np.linspace(0, 2, 101)		# time, in hours
tau = 1							# decay time, in hours
A = 1
# Generate model data.
w = A * (np.exp(-time/tau) - 1 + time/tau)

# Create figure and plot data.
plt.figure()
plt.plot(time, w, 'g-', linewidth=2)
# Add title and axes.
ax = plt.gca()
ax.set_xlabel('Time in Hours', family='monospace', fontsize=24)
ax.set_ylabel('W(t)', family='monospace', fontsize=24)
ax.set_title('Enzyme Activity Over Time', fontsize=24, weight='bold')
ax.set_xticklabels(ax.get_xticks(), family='monospace', fontsize=12)
ax.set_yticklabels(ax.get_yticks(), family='monospace', fontsize=12)
plt.show()
```

```python
#%% (b) Generate multiple curves and plot on same axes.
# Define parameters.
a_values	= [1.0 , 2.0, 3.0, 4.0]
tau_values	= [0.25, 0.5, 1.0, 2.0]
num_curves	= len(a_values)

# Sanity check.
assert len(a_values) == len(tau_values),\
		"Must have same number of values for A and tau!"

# Create empty figure to plot all data sets.
plt.figure()
# Create data sets for each combination of parameters, plot, and label.
W = np.zeros((num_curves, time.size))		# empty array to store data
for i in range(num_curves):
	tau = tau_values[i]
	A = a_values[i]
	legend_string = "A = {}, $\\tau$ = {}".format(A, tau)
	W[i] = A * (np.exp(-time/tau) - 1 + time/tau)
	plt.plot(time, W[i], linewidth=2, label=legend_string)

#%% (c) Change line styles.
ax = plt.gca()				# gain access to current Axes object
lines = ax.get_lines()		# get list of line objects in plot
# Make half of lines dashed, half dotted.
num_lines = len(lines)
for i in range(num_lines):
	if i%2:					# n%2 == 0 (False) if n is even.
		plt.setp(lines[i], linestyle='--')
	else:
		plt.setp(lines[i], linestyle=':')

#%% (d) Add title, axes, and legend.
ax.set_xlabel('Time in Hours', family='monospace', fontsize=24)
ax.set_ylabel('W(t)', family='monospace', fontsize=24)
ax.set_title('Enzyme Activity Over Time', fontsize=24, weight='bold')
ax.set_xticklabels(ax.get_xticks(), family='monospace', fontsize=12)
ax.set_yticklabels(ax.get_yticks(), family='monospace', fontsize=12)
ax.legend(loc='best')
plt.show()

```
